## Step 4: Hypothesis Testing

```python
#this time keep only the team rows in the position column
with_team = league_data[league_data['position'] == 'team']

#still remove the partial rows in datacompleteness
remove_partial = with_team[with_team['datacompleteness'] != 'complete']

#keep only relevant columns that i want to test
with_team_cols = remove_partial[['gameid', 'deaths', 'visionscore']]

#create a new column from the visionscore that maps 1 if the team has a higher vision score in the game, and a 0 else
with_team_cols['higher_vs_lower'] = with_team_cols.groupby('gameid')['visionscore'].transform(lambda col: (col == col.max()).astype(int))

#keep only deaths and the new higher_vs_lower column
deaths_higher_vs_lower = with_team_cols[['deaths', 'higher_vs_lower']]

#figure to see the distribution of the observed deaths vs visionscore
#fig_observed = px.histogram(deaths_higher_vs_lower, color='higher_vs_lower', histnorm='probability', marginal='box', barmode='overlay', opacity=0.7)
#fig_observed.show(renderer='browser')

fig_observed = px.histogram(deaths_higher_vs_lower, x='deaths', color='higher_vs_lower', 
                           histnorm='probability', marginal='box', barmode='overlay', opacity=0.7,
                           title='Distribution of Deaths by Vision Score Group')
fig_observed.show(renderer='browser')

#create the observed test statistic (higher vision score deaths - lower vision score deaths
observed_group = deaths_higher_vs_lower.groupby('higher_vs_lower')['deaths'].mean()
observed_difference = observed_group.loc[1] - observed_group.loc[0]
observed_difference

#shuffle the deaths column of the dataframe
shuffled = deaths_higher_vs_lower.assign(deaths_shuffled=np.random.permutation(deaths_higher_vs_lower['deaths']))
shuffled.head()

#permutation test 
n_reps = 1000

differences = []
for _ in range(n_reps):
    
    deaths_shuffled = deaths_higher_vs_lower.assign(deaths_shuffled=np.random.permutation(deaths_higher_vs_lower['deaths']))
    group_means = (deaths_shuffled.groupby('higher_vs_lower').mean().loc[:, 'deaths_shuffled'])
    difference = group_means.loc[1] - group_means.loc[0]
    differences.append(difference)

#first 10 rows of the shuffled dataset
differences[:10]

#create a new histogram and plot the observed difference
new_fig = px.histogram(pd.DataFrame(differences), x=0, nbins=50, histnorm='probability', 
    title='Empirical Distribution of the Mean Differences <br> in Deaths (Higher Vision Score - Lower Vision Score)')
new_fig.add_vline(x=observed_difference, line_color='orange', line_width=2, opacity=1)
new_fig.update_layout(xaxis_range=[-1.5, 1.5], margin=dict(t=60))
new_fig.show(renderer='browser')

#find the p_val where the difference is less than observed difference
p_val = np.mean(np.array(differences) <= observed_difference)
p_val
```

## Step 6: Baseline Model

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, QuantileTransformer, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.metrics import f1_score, accuracy_score, classification_report
```

```python
#baseline features: kills, deaths, assists, visionscore (quantitative), and position (categorical)
X = remove_team[['kills', 'visionscore', 'position']]
y = remove_team['result']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1, stratify=y)

preprocessor = ColumnTransformer(transformers=[
        ('quant', StandardScaler(), ['kills', 'visionscore']),
        ('cat', OneHotEncoder(drop='first', sparse_output=False), ['position'])])

base_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(random_state=1, max_iter=1000))])

base_pipeline.fit(X_train, y_train)

result_pred_test = base_pipeline.predict(X_test)

result_pred_accuracy = accuracy_score(y_test, result_pred_test)
result_pred_f1_score = f1_score(y_test, result_pred_test, average='macro')

print("=" * 50)
print("BASELINE MODEL PERFORMANCE (Test Set)")
print("=" * 50)
print(f"Test Accuracy: {result_pred_accuracy:.4f}")
print(f"Test F1-Score (macro): {result_pred_f1_score:.4f}")
print("\nClassification Report:")
print(classification_report(y_test, result_pred_test, target_names=['Loss', 'Win']))

y_pred_train = base_pipeline.predict(X_train)
train_accuracy = accuracy_score(y_train, y_pred_train)
train_f1 = f1_score(y_train, y_pred_train, average='macro')

print("\n" + "=" * 50)
print("TRAINING SET PERFORMANCE (Sanity Check)")
print("=" * 50)
print(f"Train Accuracy: {train_accuracy:.4f}")
print(f"Train F1-Score (macro): {train_f1:.4f}")

# If train >> test, you're overfitting!
if train_accuracy - result_pred_accuracy > 0.05:
    print("\n⚠️  Warning: Possible overfitting detected")
else:
    print("\n✓ Model generalizes well")
```

## Step 7: Final Model

```python
#============================================================
# STEP 7: FINAL MODEL WITH FEATURE ENGINEERING
# ============================================================

# First, let's create the engineered features
# KDA = (kills + assists) / deaths (with handling for 0 deaths)
# VisionScore per minute = visionscore / gamelength
# Total Gold per minute = totalgold / gamelength

# Create feature-engineered dataset
X_final = remove_team[['kills', 'assists', 'deaths', 'visionscore', 
                        'totalgold', 'gamelength', 'position']].copy()

# Engineer new features
X_final['kda'] = (X_final['kills'] + X_final['assists']) / X_final['deaths'].replace(0, 1)
X_final['vision_per_min'] = X_final['visionscore'] / (X_final['gamelength'] / 60)
X_final['gold_per_min'] = X_final['totalgold'] / (X_final['gamelength'] / 60)

feature_cols = ['kills', 'visionscore', 'kda', 'vision_per_min', 'gold_per_min', 'position']
X_final_selected = X_final[feature_cols]
y_final = remove_team['result']

X_train_final, X_test_final, y_train_final, y_test_final = train_test_split(
    X_final_selected, y_final, test_size=0.25, random_state=1, stratify=y_final
)

print("HYPERPARAMETER TUNING PLAN")
print("=" * 70)
print("\nHyperparameters to tune:")
print("  1. Regularization strength (C): Controls overfitting")
print("     - Lower C = stronger regularization = simpler model")
print("     - Higher C = weaker regularization = more complex model")
print("     - Testing: [0.01, 0.1, 1, 10, 100]")
print("\n  2. Penalty type: L1 (Lasso) vs L2 (Ridge)")
print("     - L1: Feature selection (sets some coefficients to 0)")
print("     - L2: Shrinks all coefficients (default for LogisticRegression)")
print("     - Testing: ['l1', 'l2']")
print("\n  3. Solver: Algorithm for optimization")
print("     - 'liblinear': Good for small datasets, supports L1")
print("     - 'saga': Supports both L1 and L2, good for large datasets")
print("     - Testing: ['liblinear', 'saga']")
print("\nRationale: These hyperparameters control model complexity and can")
print("prevent overfitting, especially with our engineered features.")
print("=" * 70)

# ============================================================
# HYPERPARAMETER TUNING WITH GRIDSEARCHCV
# ============================================================

# Define preprocessing for final model
preprocessor_final = ColumnTransformer(transformers=[
    ('quant', StandardScaler(), ['kills', 'visionscore', 'kda', 
                                   'vision_per_min', 'gold_per_min']),
    ('cat', OneHotEncoder(drop='first', sparse_output=False), ['position'])
])

# Create pipeline
final_pipeline = Pipeline([
    ('preprocessor', preprocessor_final),
    ('classifier', RandomForestClassifier(random_state=1))
])

# Define hyperparameter grid
param_distributions = {
    'classifier__n_estimators': [50, 100, 200],
    'classifier__max_depth': [5, 10, 15, None],
    'classifier__min_samples_split': [2, 5, 10],
    'classifier__min_samples_leaf': [1, 2, 4]
}


print("\n" + "=" * 70)
print("HYPERPARAMETER TUNING PLAN - RANDOMIZED SEARCH")
print("=" * 70)
print("\nUsing RandomizedSearchCV instead of GridSearchCV")
print("  - GridSearchCV: Tests ALL combinations (144 total)")
print("  - RandomizedSearchCV: Randomly samples N combinations")
print(f"  - We'll test 20 random combinations (instead of all 144)")
print("\nThis is much faster while still finding good hyperparameters!")
print("=" * 70)

# Perform grid search with cross-validation
random_search = RandomizedSearchCV(
    final_pipeline,
    param_distributions,
    n_iter=20,
    cv=5,
    scoring='f1_macro',
    n_jobs=-1,
    verbose=2,
    random_state=1 
)

print("\nPerforming RandomizedSearchCV...")
print(f"Testing 20 random combinations out of 144 possible")
print(f"Total fits: 20 combinations × 5 folds = 100 fits")
print("-" * 70)

random_search.fit(X_train_final, y_train_final)

print("\n" + "=" * 70)
print("RANDOMIZED SEARCH RESULTS")
print("=" * 70)
print(f"Best parameters: {random_search.best_params_}")
print(f"Best cross-validation F1-score: {random_search.best_score_:.4f}")

# Get the best model
best_final_model = random_search.best_estimator_

# ============================================================
# FINAL MODEL EVALUATION
# ============================================================

# Predictions on test set
y_pred_test_final = best_final_model.predict(X_test_final)
test_accuracy_final = accuracy_score(y_test_final, y_pred_test_final)
test_f1_final = f1_score(y_test_final, y_pred_test_final, average='macro')

# Predictions on train set (for overfitting check)
y_pred_train_final = best_final_model.predict(X_train_final)
train_accuracy_final = accuracy_score(y_train_final, y_pred_train_final)
train_f1_final = f1_score(y_train_final, y_pred_train_final, average='macro')

print("\n" + "=" * 70)
print("FINAL MODEL PERFORMANCE (Test Set)")
print("=" * 70)
print(f"Test Accuracy: {test_accuracy_final:.4f}")
print(f"Test F1-Score (macro): {test_f1_final:.4f}")
print("\nClassification Report:")
print(classification_report(y_test_final, y_pred_test_final, target_names=['Loss', 'Win']))

print("\n" + "=" * 70)
print("TRAINING SET PERFORMANCE (Sanity Check)")
print("=" * 70)
print(f"Train Accuracy: {train_accuracy_final:.4f}")
print(f"Train F1-Score (macro): {train_f1_final:.4f}")

if train_accuracy_final - test_accuracy_final > 0.05:
    print("\n⚠️  Warning: Possible overfitting detected")
else:
    print("\n✓ Model generalizes well")

rf_model = best_final_model.named_steps['classifier']
feature_names_after_preprocessing = (
    ['kills', 'visionscore', 'kda', 'vision_per_min', 'gold_per_min'] +
    list(best_final_model.named_steps['preprocessor']
         .named_transformers_['cat']
         .get_feature_names_out(['position']))
)

importances = rf_model.feature_importances_
feature_importance_df = pd.DataFrame({
    'feature': feature_names_after_preprocessing,
    'importance': importances
}).sort_values('importance', ascending=False)

print("\n" + "=" * 70)
print("FEATURE IMPORTANCES")
print("=" * 70)
print(feature_importance_df.to_string(index=False))

# ============================================================
# COMPARISON WITH BASELINE
# ============================================================

print("\n" + "=" * 70)
print("MODEL COMPARISON: BASELINE vs FINAL")
print("=" * 70)
print(f"{'Metric':<25} {'Baseline':<15} {'Final':<15} {'Improvement':<15}")
print("-" * 70)
print(f"{'Test Accuracy':<25} {result_pred_accuracy:<15.4f} {test_accuracy_final:<15.4f} {test_accuracy_final - result_pred_accuracy:+.4f}")
print(f"{'Test F1-Score (macro)':<25} {result_pred_f1_score:<15.4f} {test_f1_final:<15.4f} {test_f1_final - result_pred_f1_score:+.4f}")
print("=" * 70)

if test_f1_final > result_pred_f1_score:
    print("✓ Final model successfully improved over baseline!")
else:
    print("⚠️  Final model did not improve over baseline")
    
improvement_pct = ((test_f1_final - result_pred_f1_score) / result_pred_f1_score) * 100
print(f"\nRelative improvement: {improvement_pct:+.2f}%")
```

## Step 8: Fairness Analysis

```python
# i will analyze across positions (support players vs. other positions)
# group x: support players, group y: other positions (top, jungle, mid, bot)

# this is because support usually has fewer kills, less gold, fewer deaths, but more assists, and higher vision score. i want to test
# if the model can perform with support vs. the other positions 

# Null hypothesis: 
# The model is fair. Its precision for support and non-support players is around the same, and differences are due to random chance.

# Alternative Hypothesis:
# The model is unfair. Its precision for support players is not the same as its precision for non-support players.

# Add position column to test set for grouping
X_test_with_position = X_test_final.copy()
y_test_with_predictions = y_test_final.copy()
y_test_with_predictions = pd.DataFrame({
    'true_result': y_test_final.values,
    'pred_result': y_pred_test_final,
    'position': X_test_final['position'].values
})

# Create binary groups
y_test_with_predictions['is_support'] = (y_test_with_predictions['position'] == 'sup').astype(int)

print("\nGroups for Fairness Analysis:")
print("  Group X: Support players (position = 'sup')")
print("  Group Y: Non-support players (position = 'top', 'jng', 'mid', 'bot')")
print("\nRationale: Supports have different playstyles (fewer kills, more vision),")
print("so we want to check if our model predicts equally well for both groups.")

# ============================================================
# CHOOSE EVALUATION METRIC
# ============================================================

print("\n" + "-" * 70)
print("Evaluation Metric: Precision")
print("-" * 70)
print("Precision = True Positives / (True Positives + False Positives)")
print("In our context: Of all games we predicted as WINS, what % were actual wins?")
print("\nWhy precision?")
print("  - If our model has lower precision for supports, it means we're")
print("    falsely predicting wins for supports more often than other roles")
print("  - This could unfairly misrepresent support player performance")

# ============================================================
# COMPUTE OBSERVED PRECISION FOR EACH GROUP
# ============================================================

from sklearn.metrics import precision_score

# Group X: Support players
support_mask = y_test_with_predictions['is_support'] == 1
y_true_support = y_test_with_predictions.loc[support_mask, 'true_result']
y_pred_support = y_test_with_predictions.loc[support_mask, 'pred_result']

# Group Y: Non-support players  
non_support_mask = y_test_with_predictions['is_support'] == 0
y_true_non_support = y_test_with_predictions.loc[non_support_mask, 'true_result']
y_pred_non_support = y_test_with_predictions.loc[non_support_mask, 'pred_result']

# Calculate precision for each group
precision_support = precision_score(y_true_support, y_pred_support, zero_division=0)
precision_non_support = precision_score(y_true_non_support, y_pred_non_support, zero_division=0)

observed_difference = precision_support - precision_non_support

print("\n" + "=" * 70)
print("OBSERVED PRECISION")
print("=" * 70)
print(f"Support players (n={support_mask.sum()}):     Precision = {precision_support:.4f}")
print(f"Non-support players (n={non_support_mask.sum()}): Precision = {precision_non_support:.4f}")
print(f"\nObserved difference (Support - Non-support): {observed_difference:+.4f}")

# ============================================================
# PERMUTATION TEST
# ============================================================

print("\n" + "=" * 70)
print("PERMUTATION TEST")
print("=" * 70)

print("\nNull Hypothesis:")
print("  Our model is fair. Its precision for support and non-support players")
print("  are roughly the same, and any differences are due to random chance.")

print("\nAlternative Hypothesis:")
print("  Our model is unfair. Its precision for support players is different")
print("  from its precision for non-support players.")

print("\nTest Statistic: Difference in precision (Support - Non-support)")
print(f"Observed test statistic: {observed_difference:.4f}")

# Perform permutation test
n_permutations = 1000
permuted_differences = []

print(f"\nRunning {n_permutations} permutations...")

np.random.seed(42)

for i in range(n_permutations):
    # Shuffle the group labels (is_support)
    shuffled_groups = y_test_with_predictions['is_support'].sample(frac=1, replace=False).values
    
    # Recalculate precision for shuffled groups
    shuffled_support_mask = shuffled_groups == 1
    shuffled_non_support_mask = shuffled_groups == 0
    
    # Get predictions for shuffled groups
    y_true_shuffled_support = y_test_with_predictions.loc[shuffled_support_mask, 'true_result']
    y_pred_shuffled_support = y_test_with_predictions.loc[shuffled_support_mask, 'pred_result']
    
    y_true_shuffled_non_support = y_test_with_predictions.loc[shuffled_non_support_mask, 'true_result']
    y_pred_shuffled_non_support = y_test_with_predictions.loc[shuffled_non_support_mask, 'pred_result']
    
    # Calculate precision for shuffled groups
    precision_shuffled_support = precision_score(y_true_shuffled_support, y_pred_shuffled_support, zero_division=0)
    precision_shuffled_non_support = precision_score(y_true_shuffled_non_support, y_pred_shuffled_non_support, zero_division=0)
    
    # Store the difference
    permuted_difference = precision_shuffled_support - precision_shuffled_non_support
    permuted_differences.append(permuted_difference)
    
    # Progress indicator
    if (i + 1) % 200 == 0:
        print(f"  Completed {i + 1}/{n_permutations} permutations...")

permuted_differences = np.array(permuted_differences)

# Calculate p-value (two-tailed test)
p_value = np.mean(np.abs(permuted_differences) >= np.abs(observed_difference))

print("\n" + "=" * 70)
print("PERMUTATION TEST RESULTS")
print("=" * 70)
print(f"Observed difference: {observed_difference:+.4f}")
print(f"P-value: {p_value:.4f}")
print(f"\nSignificance level: 0.05")

if p_value < 0.05:
    print("\n⚠️  CONCLUSION: REJECT the null hypothesis")
    print("    Our model appears to be UNFAIR. The difference in precision between")
    print("    support and non-support players is statistically significant.")
    print(f"    Support players have {'higher' if observed_difference > 0 else 'lower'} precision.")
else:
    print("\n✓ CONCLUSION: FAIL TO REJECT the null hypothesis")
    print("    Our model appears to be FAIR. The difference in precision between")
    print("    support and non-support players is likely due to random chance.")

# ============================================================
# VISUALIZATION
# ============================================================

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(permuted_differences, bins=50, edgecolor='black', alpha=0.7)
plt.axvline(observed_difference, color='red', linestyle='--', linewidth=2, 
            label=f'Observed difference: {observed_difference:.4f}')
plt.axvline(-observed_difference, color='red', linestyle='--', linewidth=2)
plt.xlabel('Difference in Precision (Support - Non-support)')
plt.ylabel('Frequency')
plt.title('Permutation Test: Distribution of Differences Under Null Hypothesis')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\n" + "=" * 70)
print("FAIRNESS ANALYSIS COMPLETE")
print("=" * 70)
```
